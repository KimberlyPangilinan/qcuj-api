```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import csv

# Install Webdriver
service = Service(ChromeDriverManager().install())

# Create Driver Instance
driver = webdriver.Chrome(service=service)

page = 1
baseUrl = f'https://www.mdpi.com/search?sort=pubdate&page_no={page}&journal=education&view=default'


driver.get(baseUrl)

driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

time.sleep(2)

soup = BeautifulSoup(driver.page_source, 'html5lib')

# items = soup.findAll('div', {'class':'generic-item article-item'})

titles = soup.findAll('a',{'class':'title-link'})
abstracts = soup.findAll('div',{'class':'abstract-full'})
journal_id = 2
# titles2 = [title.text for title in titles]
# abstracts2 = [abstract.text.strip() for abstract in abstracts]

finalTitleAbstract = []
for i in range(len(titles)):
    finalTitleAbstract.append([titles[i].text, abstracts[i].text.strip(),journal_id])

# finalTitle = titles2.copy()
# finalAbstracts = abstracts2.copy()

# finalTitleAbstract = titleAbstract.copy()

max_page = int(soup.find('div', {'class':'columns large-6 medium-6 small-12'}).text.split(' ')[-1].strip()[:-1])

print(f"Page {page} {len(finalTitleAbstract)}")

for page in range(2,max_page):
    baseUrl = f'https://www.mdpi.com/search?sort=pubdate&page_no={page}&journal=education&view=default'
    driver.get(baseUrl)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    time.sleep(2)

    soup = BeautifulSoup(driver.page_source, 'html5lib')

    titles = soup.findAll('a',{'class':'title-link'})
    abstracts = soup.findAll('div',{'class':'abstract-full'})

    for i in range(len(titles)):
        finalTitleAbstract.append([titles[i].text, abstracts[i].text.strip(),journal_id])

    # titles2 = [title.text for title in titles]
    # abstracts2 = [abstract.text.strip() for abstract in abstracts]

    # finalTitle.extend(titles2)
    # finalAbstracts.extend(abstracts2)

    # finalTitleAbstract.append(titleAbstract)

    print(f"Page {page} {len(finalTitleAbstract)}")

    # if len(finalTitleAbstract) >= 150: break


fields = ['title','abstract','journal_id']
with open('data/mdpi.csv','w', encoding="utf-8") as file:
    csvwriter = csv.writer(file)

    ## write header fields
    csvwriter.writerow(fields)

    ## write values
    for i,row in enumerate(finalTitleAbstract):
        csvwriter.writerow(row)
        print(i)

driver.close()
```

```python
import csv
count_gavel = 0
count_lamp = 0
count_star = 0
with open('data/mdpi3.csv', mode ='r',encoding="utf-8")as file:
  csvFile = csv.reader(file)
  next(csvFile)
  for i,lines in enumerate(csvFile):
        if len(lines) == 0: continue
        if(int(lines[2])-1 == 0): count_gavel += 1
        if(int(lines[2])-1 == 1): count_lamp += 1
        if(int(lines[2])-1 == 2): count_star += 1
        
print(count_gavel)
print(count_lamp)
print(count_star)
```
